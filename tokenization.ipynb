# TOKENIZATION-converting sentence to single words(The main goal is to keep vocabulary as small as possible.(So we create tokens here)


```python
import csv
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
with open('Allheadlines.csv') as csvfile:
    readCSV = csv.reader(csvfile)
    for row in readCSV:
        example_text = row[0]
        tokenized = word_tokenize(example_text)
        print(tokenized)
```
