# Stop words and converting all text to lower case- filtering unneccessary words in the tokenized word .The main goal of converting all text to lower case is that in natural language processing Accident and accident are two different things. This process is called normalisation


```python
import csv
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
with open('Allheadlines.csv') as csvfile:
    readCSV = csv.reader(csvfile)
    for row in readCSV:
        example_text = row[0]
        All_lowercase = example_text.lower()
        tokenized = word_tokenize(All_lowercase)
        stop_words = set(stopwords.words("English"))
        filtered_sentence = [w for w in tokenized if not w in stop_words]
        print(filtered_sentence)
        #for w in tokenized:
            #if w not in stop_words:
                #filtered_sentence.append(w)
                #print(filtered_sentence)
```
